```python
def plot_particle_cloud(particles, ax=None, lm=0.3, c="red"):
    plt.scatter(particles[:, 1], particles[:, 2], s=100*np.sqrt(particles[:, 0])+1, alpha=0.5, c=c)
    plt.xlabel("eta")
    plt.ylabel("phi")
    plt.xlim(-lm, lm)
    plt.ylim(-lm, lm)
    return ax

plt.figure()
plot_particle_cloud(batch[0][0].numpy())
```
