```python
df_sig = df[df["is_signal"]==True]
df_bkg = df[df["is_signal"]==False]
```

```python
plt.figure(figsize=(8, 2))
bins = np.linspace(3000, 4600, 100)
plt.axvline(x=3000, color='black', linestyle='--')
plt.axvline(x=4600, color='black', linestyle='--')
plt.axvline(x=3300, color='black', linestyle='-')
plt.axvline(x=3700, color='black', linestyle='-')
plt.axvspan(3000, 3300, color='yellow', alpha=0.2)
plt.axvspan(3700, 4600, color='yellow', alpha=0.2, label="Sidebands")
plt.axvspan(3300, 3700, color='lime', alpha=0.2, label="Signal region")
plt.hist([df_bkg["m_jj"], df_sig["m_jj"][:10000]], bins=bins, stacked=True, label=['background', 'signal'], color=['gray', 'red'])
#plt.legend(loc='upper right', framealpha=1.0)
#plt.xlabel("$m_{JJ}$")
#plt.xlabel("$m_{JJ}$")
plt.xticks([])
plt.yticks([])
```
